```python
def pegarValorPdf(df,banco,index,colunaNome):
    return df[df[1].str.upper().str.contains(banco, case=False, na=False,)][index].astype(float).sum(numeric_only=True)
```

```python
novaMemoria = r"C:\Users\jcampbell1\OneDrive - SEFAZ-RJ\Folha EGE\Exercício 2024\Teste - Marinete\Memória de Cálculo - " + str(date.today().month) + "." + str(date.today().year) + ".xlsx"


planilha = load_workbook(novaMemoria)
sequencial = planilha["Sequencial"]

celulas  ={
    'COMPLEMENTO SALARIO MINIMO FEDERAL' :  sequencial["F36"],
    'AUXÍLIO ADOÇÃO' : sequencial['F40'],
    "3190.92.00": sequencial["F39"]
}

df = tabula.read_pdf(r"C:\Users\jcampbell1\Downloads\TGRJ0807P_Agosto.pdf", pages='all', pandas_options={'header': None})[0]


index = 3
df[index] = df[index].str.replace('.', '')
df[index] = df[index].str.replace(',', '.')


 
for nome,banco in celulas.items():
        banco.value = pegarValorPdf(df,nome,index)

planilha.save(novaMemoria)




```
